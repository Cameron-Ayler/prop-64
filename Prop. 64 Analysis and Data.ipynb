We can use the merge() method to combine to dataframes into one.

The dataframes must have some type of commmonality with which they can be combined. This called a "key".

In relational databases, when combining two dataframes, the smaller one will be incorporated into the 
larger one, eliminating all the data in the larger one that does not correlate to the smaller one. In this 
way, the smaller dataset acts as a "filter" of sorts.

In our example, we will see how many of the 56,379 contributions (contrib) are connected to Proposition 64
committees (weed_prop). The contributions will be filtered down based on how many of them were to Prop 64
committees. The commonality between them ("key") is the calaccess_committee_id column. We will set them to
a new variable (weed_contribs) in order to save the new dataframe.

The merge() function first takes as arguments the dataframes being used, followed by optional inputs. In this
case, we will use the "on" argument, which allows us to specify the "key" (column) to use for joining.

```python
pd.merge(contrib, weed_prop, on = "calaccess_committee_id")
```

```python
weed_contribs = pd.merge(contrib, weed_prop, on = "calaccess_committee_id")
```

```python
weed_contribs.info()
```

```python
# Out of the the 56,379 total contributions, 860 of them were to committees involved with Prop 64.
# In addition, the new combined dataframe (weed_contribs) has 20 columns, as opposed to the 13 from contrib
# and 8 from weed_prop. calaccess_committee_id column is only shown once, meaning either contrib only
# puts in 12 columns, or weed_prop only puts in 7. Since both dataframes had a committee_name column, they
# have been differentiated in the new dataframe by adding x's and y's to the end of their names (committee_name_x
# and committee_name_y)
```
